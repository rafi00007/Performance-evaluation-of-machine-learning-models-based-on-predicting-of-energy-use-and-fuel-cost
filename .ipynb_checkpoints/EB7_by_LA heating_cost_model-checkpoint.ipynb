```python
import pandas as pd
dataset= pd.read_csv(r"C:\Users\rafi\Desktop\master pro\EB7_by_LA2.csv", encoding= 'unicode_escape', delimiter= ',')
print(dataset)
```

```python
dataset.dtypes
```

```python
dataset['EnergyUseAveragePerDwelling'] = dataset['EnergyUseAveragePerDwelling'].astype(int)
dataset['AverageFloorArea'] = dataset['AverageFloorArea'].astype(int)
dataset['CarbonDioxideEmissionsAveragePerDwelling'] = dataset['CarbonDioxideEmissionsAveragePerDwelling'].astype(int)
dataset['HeatingCostAveragePerDwelling'] = dataset['HeatingCostAveragePerDwelling'].astype(int)
del dataset['CarbonDioxideEmissionsAveragePerDwelling']
del dataset['LightingCostAveragePerDwelling']
del dataset['HotWaterCostAveragePerDwelling']
```

```python
from sklearn.model_selection import train_test_split
X=dataset[['AverageFloorArea','EnergyUseAveragePerDwelling']]
Y=dataset['HeatingCostAveragePerDwelling']
X_train, X_test, Y_train, Y_test =train_test_split(X,Y, test_size=0.2, random_state = 0)
X
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(X_train)
X_train_scaled= scaler.transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

## CNN

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np

 
# Defining the Input layer and FIRST hidden layer, both are same!
Sample_size=X_test_scaled.shape[0]
time_steps= X_test_scaled.shape[1]
input_dim=1
train_data_reshaped=X_test_scaled.reshape(Sample_size,time_steps,input_dim)
print(train_data_reshaped.shape)
def build_conv1D_model():
    n_timesteps=train_data_reshaped.shape[1]
    n_feature=train_data_reshaped.shape[2]
    model= keras.Sequential(name='model_conv1D')
    model.add(keras.layers.Input(shape=(n_timesteps,n_feature)))
    model.add(keras.layers.Conv1D(filters=64, kernel_size=1, activation='relu', name='Conv1D_1'))
    model.add(keras.layers.Dropout(0.5))
    model.add(keras.layers.Conv1D(filters=32, kernel_size=1, activation='relu', name='Conv1D_2'))
    #model.add(keras.layers.MaxPooling1D(pool_size=1,name="maxpooling_1"))
    model.add(keras.layers.Conv1D(filters=16, kernel_size=1, activation='relu', name='Conv1D_4'))
    model.add(keras.layers.MaxPooling1D(pool_size=1,name="maxpooling_2"))
    model.add(keras.layers.Flatten())
    model.add(keras.layers.Dense(32, activation='linear',name='Dense_1'))
    #model.add(keras.layers.Dense(32, activation='relu',name='Dense_2'))
    model.add(keras.layers.Dense(n_feature, activation='linear',name='Dense_3'))
    
              
    optimizer=tf.keras.optimizers.RMSprop(0.001)
    model.compile(loss='mse',optimizer=optimizer, metrics=['mae'])
    return model
    
model=build_conv1D_model()
model.summary()
```

```python
# Fitting the ANN to the Training set
model.fit(train_data_reshaped,Y_train ,epochs = 4000,validation_split=0.2)
```

```python
from sklearn.metrics import r2_score
r2_score(Y_test, np.squeeze(model.predict(train_data_reshaped)) )
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
mse= mean_squared_error(Y_test, np.squeeze(model.predict(train_data_reshaped)) )
mae= mean_absolute_error(Y_test, np.squeeze(model.predict(train_data_reshaped)) )
print('mean squired error:', mse)
print('mean absolute error:', mae)
```

```python
import matplotlib.pyplot as plt
plt.scatter(Y_test, np.squeeze(model.predict(train_data_reshaped)) )
plt.xlabel('Actual')
plt.ylabel('predicted')
plt.title("Actual vs predicted")
```

```python
y_prd=np.squeeze(np.squeeze(model.predict(train_data_reshaped)) )
predic_dif=pd.DataFrame({'Actual':Y_test,'predict':y_prd,"differences":Y_test-y_prd})
predic_dif[0:15]
```

## CNN keras tuner

```python
from tensorflow import keras
from keras import layers
from keras.layers import Dense
from kerastuner.tuners import RandomSearch
def build_model(hp):
    model2= keras.Sequential()
    for i in range(hp.Int('num_layers',2,20)):
        model2.add(layers.Dense(units=hp.Int('units_' + str(i),
                                           min_value=32,
                                           max_value=512,
                                           step=32),
                              activation='relu'))
        model2.add(layers.Dense(1, activation='linear',name='Dense_1'))
        model2.compile(
              optimizer=keras.optimizers.Adam(
                 hp.Choice('learning rate', [1e-2,1e-3,1e-4])),
              loss='mean_absolute_error',
              metrics=['mean_absolute_error'])
        return model2
```

```python
tuner=RandomSearch(build_model,objective='val_mean_absolute_error',max_trials=50, executions_per_trial=3,directory='pproject1',project_name='ABC')
```

```python
tuner.search(X_train_scaled,Y_train ,epochs=100,validation_data=(X_test_scaled,Y_test))
```

```python
best_hps=tuner.get_best_hyperparameters(num_trials=1)[0]
```

```python
model4 = tuner.hypermodel.build(best_hps)
model4.fit(X_train_scaled, Y_train, epochs=10000, validation_split=0.2)
```

```python
import numpy as np
from sklearn.metrics import r2_score
r2_score(Y_test, np.squeeze(model4.predict(X_test_scaled)) )
```

```python
prediction =np.squeeze(model4.predict([[80,220]]))
print("Prediction: {}".format(prediction))
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
mse= mean_squared_error(Y_test, np.squeeze(model4.predict(X_test_scaled)) )
mae= mean_absolute_error(Y_test, np.squeeze(model4.predict(X_test_scaled)) )
print('mean squired error:', mse)
print('mean absolute error:', mae)
```

```python
import matplotlib.pyplot as plt
plt.scatter(Y_test, np.squeeze(model4.predict(X_test_scaled)) )
plt.xlabel('Actual')
plt.ylabel('predicted')
plt.title("Actual vs predicted")
```
